```python
def unifrac(c1, c2, tree):

    # count number of unique and total branches
    unique = 0
    total = 0

    # loop over nodes in phylogenetic tree
    for el in tree:

        # any shared elements between node and comm1
        # elinC1 = any(np.in1d(el, c1))
        elinC1 = bool(set(el) & set(c1))
        
        # any shared elements between node and comm2
        # elinC2 = any(np.in1d(el, c2))
        elinC2 = bool(set(el) & set(c2))

        # exclusive OR to count unique branches
        unique += int(elinC1 ^ elinC2)

        # inclusive OR to count unique + shared (total) branches 
        total += int(elinC1 or elinC2)

    return unique / total
```

```python
# species / fiber phylogenetic tree 
tree = ({'Inulin'},
        {'Xylan'},
        {'Pectin'},
        {'Starch'},
        {'ArGal'},
        {'AcGum'},
        {'ER'}, 
        {'RI'},
        {'CC'}, 
        {'BH'}, 
        {'DF'},
        {'AC'},
        {'FP'},
        {'CH'}, 
        {'CA'}, 
        {'EL'}, 
        {'BA'},
        {'BL'}, 
        {'PJ'}, 
        {'BH'},
        {'PC'},
        {'ER', 'RI'}, 
        {'BH', 'DF'},
        {'CA', 'EL'}, 
        {'BA', 'BL'}, 
        {'BU', 'PC'},
        {'ER', 'RI', 'CC'},
        {'PJ', 'BU', 'PC'},
        {'CA', 'EL', 'BA', 'BL'},
        {'ER', 'RI', 'CC', 'BH', 'DF'},
        {'ER', 'RI', 'CC', 'BH', 'DF', 'AC'},
        {'ER', 'RI', 'CC', 'BH', 'DF', 'AC', 'FP'},
        {'ER', 'RI', 'CC', 'BH', 'DF', 'AC', 'FP', 'CH'},
        {'ER', 'RI', 'CC', 'BH', 'DF', 'AC', 'FP', 'CH', 'CA', 'EL', 'BA', 'BL'})
```

```python
# make sure same comm is zero
c1 = {'ER'}
c2 = {'ER'}

unifrac(c1, c2, tree)
```

```python
# most different individual species
c1 = {'ER'}
c2 = {'PC'}

unifrac(c1, c2, tree)
```

```python
# less different individual species
c1 = {'ER'}
c2 = {'AC'}

unifrac(c1, c2, tree)
```

```python
# similar community
c1 = {'AC', 'BU', 'ER'}
c2 = {'AC', 'BU', 'RI'}

unifrac(c1, c2, tree)
```

```python
# more different community 
c1 = {'AC', 'BU', 'ER'}
c2 = {'AC', 'BU', 'CC'}

unifrac(c1, c2, tree)
```

```python
# different community with different fibers
c1 = {'AC', 'BU', 'ER', 'Inulin'}
c2 = {'AC', 'BU', 'CC', 'Xylan'}

unifrac(c1, c2, tree)
```

```python
# make sure triangle inequality is satisfied
# d(A,C) ≤ d(A,B) + d(B,C)

A = {'AC', 'Inulin'}
B = {'AC', 'ER', 'CC', 'CH', 'Inulin'}
C = {'AC', 'PJ', 'BH', 'Inulin', 'Xylan', 'AcGum'}

unifrac(A, C, tree) <= (unifrac(A, B, tree) + unifrac(B, C, tree))
```
