Roughness-slope ratio 

- Nonlinearity does not necessarily mean rugged, r/s can overestimate ruggedness of functions that are nonlinear but not rugged 

$$ f(x) = x^2$$ 

- r/s will overestimate ruggedness for functions that depend weakly on inputs 

$$ g(x) = x_1 + .001 x_2 $$ 

will be considered more rugged than 

$$ g(x) = x_1 + x_2 $$
